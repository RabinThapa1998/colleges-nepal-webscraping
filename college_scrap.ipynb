```python

# Specify the Chrome version explicitly
chrome_version = "126.0.6478.126"  # Replace with your Chrome version
service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

# URL with infinite scroll
# url ="https://www.collegesnepal.com/college/"
# Open the webpage
driver.get(url)

# Scroll down to the bottom of the page to fetch all content
last_height = driver.execute_script("return document.body.scrollHeight")
i=0
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for new content to load
    time.sleep(10)  # Adjust sleep time as needed
    
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    print(i)
    i = i+1

# Get the page source after loading all content
page_content = driver.page_source

# Parse the page content with BeautifulSoup
soup = BeautifulSoup(page_content, 'html.parser')

# Find the <ul> with the specific class and ID
ul_element = soup.find('ul', class_='college_b mt-3', id='load_data')

if ul_element:
    # Find all <li> elements within the <ul>
    li_elements = ul_element.find_all('li')
    
    # Process each <li> element
    for li in li_elements:
        # Extract and process data as needed
        print(li.text.strip())
else:
    print('No <ul> element found with class="college_b mt-3" and id="load_data"')

# Close the WebDriver
driver.quit()
```

```python

contact_list = []
 # Iterate over each <li> element
for li in li_elements:
    # Find the <a> element within the <li>
    a_element = li.find('a')
    if a_element:
        # Get the href attribute (link)
        link = a_element['href']
        # Print the link (optional)
        print(f'Processing link: {link}')

        # Send a request to the link
        link_response = requests.get(link)

        # Parse the link content with BeautifulSoup
        link_soup = BeautifulSoup(link_response.content, 'html.parser')

          # Find the 'Contact' tab
        contact_tab = None
        nav_items = link_soup.find_all('li', class_='nav-item')

        for item in nav_items:
            if item.find('a') and item.find('a').text.strip() == 'Contact':
                contact_tab = item.find('a')['href']
                break

        # Print the result
        print(contact_tab)
        contact_list.append(link)


```

```python
# Initialize an empty list to store all contact info
all_contact_info = []

# Specify the Chrome version explicitly
chrome_version = "126.0.6478.126"  # Replace with your Chrome version
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Iterate over each URL in the contact_list
for url in contact_list:
    try:
        # Open the webpage
        driver.get(url)
        # Find the "Contact" tab using its text
        contact_tab = driver.find_element(By.LINK_TEXT, 'Contact')
        # Click the "Contact" tab
        contact_tab.click()
        # Optional: Wait for the page to load fully
        time.sleep(2)
        # Get the page content after clicking the "Contact" tab
        page_content = driver.page_source
        # Parse the new page content with BeautifulSoup
        soup = BeautifulSoup(page_content, 'html.parser')
        # Find contact_info
        contact_info_div = soup.find(class_="contact_info")
        cname = contact_info_div.find(class_="cname").text.strip()
        
        # Extract phone number
        phone = contact_info_div.find("i", class_="fa-phone").next_sibling.strip()
        
        # Extract mobile information (website and email in this case)
        mobile_info = contact_info_div.find_all("i", class_="fa-mobile")
        website = mobile_info[0].next_sibling.strip()
        email = mobile_info[1].next_sibling.strip()
        
        contact_info = {
            "cname": cname,
            "phone": phone,
            "website": website,
            "email": email,
            "url": url  # Adding the original URL to the dictionary
        }
        
        # Append the contact_info dictionary to the list
        all_contact_info.append(contact_info)
        
        # Print for debugging
        print(contact_info)
        
    except Exception as e:
        print(f'Error processing {url}: {e}')

# Close the WebDriver
driver.quit()

# Now you have all contact info stored in all_contact_info list
print(f"Total contacts processed: {len(all_contact_info)}")
```

```python
import csv

# Assuming all dictionaries have the same keys
keys = all_contact_info[0].keys()

with open('collegeScrapeData.csv', 'w', newline='') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(all_contact_info)
```
